```python
import torch
```

```python
group_size = 3
block_size = 3
x = torch.tril(torch.ones(block_size, block_size))
x = torch.repeat_interleave(x, group_size, dim=1)
x = torch.repeat_interleave(x, group_size, dim=0)
x.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(x, linewidths=0.2,
            linecolor="black", cbar=False)
plt.axis("off")
```

```python
import torch
import seaborn as sns
from hergen.models.group_causal_transformer import TemporalPositionalEncoding
```

```python
pe = TemporalPositionalEncoding(d_model=768)
```

```python
study_date = torch.randint(0, 5000, (10,))
study_date = torch.sort(study_date).values
study_date = study_date.unsqueeze(0)
tpe = pe(study_date)
tpe = tpe.squeeze()
print(study_date)
print(tpe)
sns.heatmap(tpe)
```
