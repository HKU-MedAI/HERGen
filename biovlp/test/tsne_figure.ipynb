```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cvt_df = pd.read_csv('tsne_feats_cvt.csv')
temporal_df = pd.read_csv('tsne_feats_temporal.csv')
all_diseases = cvt_df['disease'].unique()
# cvt_df = cvt_df.replace({
#     "Atelectasis": 0,
#     "Cardiomegaly": 1,
#     "Lung Opacity": 2,
#     "Pleural Effusion": 3,
#     "Pneumonia": 4,
# })

# temporal_df = temporal_df.replace({
#     "Atelectasis": 0,
#     "Cardiomegaly": 1,
#     "Lung Opacity": 2,
#     "Pleural Effusion": 3,
#     "Pneumonia": 4,
# })
```

```python
import matplotlib

# fig, (ax1, ax2) = plt.subplots(2, 1)
fig, ax = plt.subplots(1, 1)
fig.set_figwidth(8)
fig.set_figheight(5)
ax.set_axis_off()

cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=0, vmax=4)

for i, disease in enumerate(all_diseases):
    select_cvt_df = cvt_df[cvt_df['disease'] == disease]
    ax.scatter(select_cvt_df['t1'], select_cvt_df['t2'], label=disease, c=cmap(norm(i)), s=20)

ax.legend()

plt.savefig('tsne_cvt.pdf', dpi=300, bbox_inches='tight')
```
