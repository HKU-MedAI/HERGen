```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
similarity_cvt = pd.read_csv('similarities_cvt.csv')
similarity_cvt.rename(columns={
    '2->1': '1',
    '3->1': '2',
    '4->1': '3',
    '5->1': '4',
    },
    inplace=True)
similarity_temporal = pd.read_csv('similarities_temporal.csv')
similarity_temporal.rename(columns={
    '2->1': '1',
    '3->1': '2',
    '4->1': '3',
    '5->1': '4',
    },
    inplace=True)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
sns.boxplot(data=similarity_cvt, ax=ax1)
# ax1.set_ylim(-1, 1)
ax1.set_title("CvT21k2DistilGPT2", fontsize=14)
ax1.set_xlabel("Time Step", fontsize=14)
ax1.set_ylabel("Similarity", fontsize=14)

sns.boxenplot(data=similarity_temporal, ax=ax2)
# ax2.set_ylim(-1, 1)
ax2.set_title("HER2Gen (Ours)", fontsize=14)
ax2.set_xlabel("Time Step", fontsize=14)
ax2.set_ylabel("Similarity", fontsize=14)
# plt.legend()
plt.savefig("similarity.pdf", bbox_inches='tight')
```
